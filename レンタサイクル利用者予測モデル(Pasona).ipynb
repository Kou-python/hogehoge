### 予測・モデル構築
訓練⽤データセットとテスト⽤データセットに分割しましょう。単純に全体の何割かを訓練⽤データセットとし、残りをテス
ト⽤データセットとする、といった分割を⾏う⽅法はホールドアウト法 (holdout method) と呼ばれます。

#### 重回帰モデル

```python
# データセットを分割する関数の読み込み
from sklearn.model_selection import train_test_split
# 不要な説明変数を削除
x, y = dataset.drop(["id", "dteday", "cnt","atemp","weekday"], axis=1), dataset["cnt"]
# 訓練⽤データセットとテスト⽤データセットへの分割
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=0)

from sklearn.linear_model import LinearRegression
# モデルの定義
clf_model = LinearRegression()

# モデルの訓練
clf_model.fit(x_train, y_train) # 訓練データセットの説明変数と⽬的変数を引数に渡す
# LinearRegression(copy_X=True, fit_intercept=True, n_jobs=None, normalize=False)

stem_df = pd.DataFrame({"features":x_train.columns,"coef":clf_model.coef_}).sort_values(by="coef")
display(stem_df)
print("切⽚(バイアス)は{}".format(clf_model.intercept_))
```

重み

```python
plt.subplots(figsize=(15,10)) # グラフのサイズを指定
x = stem_df.features
y = stem_df.coef
x_position = range(len(x))
plt.barh(x_position, y, tick_label=x) # 横棒グラフ(barth)
plt.grid()
plt.show()
```

#### MAE

```python
# テストデータセットで推論
test_pre_lr = clf_model.predict(x_test)
# 予測した結果と実際の値を結合する
test_result_lr = pd.DataFrame({"actual": y_test, "predict":test_pre_lr}).reset_index(drop=True)
# 訓練用データセットで推論
train_pre_lr = clf_model.predict(x_train)
# 予測した結果と実際の値を結合する
train_result_lr = pd.DataFrame({"actual": y_train, "predict":train_pre_lr}).reset_index(drop=True)

from sklearn.metrics import mean_absolute_error #MAEを算出するためのライブラリ
train_mae_lr = mean_absolute_error(y_train, train_pre_lr)
print("訓練データセットのMAE: {0}".format(train_mae_lr))

test_mae_lr = mean_absolute_error(y_test, test_pre_lr)
print("テストデータセットのMAE: {0}".format(test_mae_lr))
```

```python
plt.subplots(figsize=(15,10))
x = test_result_lr["actual"]
y = test_result_lr["predict"]
plt.scatter(x, y)
plt.plot([0, 8800], [0, 8800], '-')
plt.xlabel("actual")
plt.ylabel("predict")
plt.xlim(0,8800)
plt.ylim(0,8800)
plt.grid()
plt.show()
```

## *重回帰モデル*

前処理

```python
import numpy as np
df=dataset
df = df[df['cnt'] != 22]
df = df[df["windspeed"] < 0.41]

df['windspeed'].fillna(df["windspeed"].median(), inplace=True)
df= df.drop(["id", "dteday", "atemp", "weekday", "hum", "holiday"], axis=1)

#ダミー変数化
test_d = df[["season", "mnth", "weathersit"]]
df=df.drop(["season", "mnth", "weathersit"], axis=1)
dummy1 = pd.get_dummies(test_d, columns=["season"], prefix=["season"], drop_first=True)
dummy2 = pd.get_dummies(test_d, columns=["mnth"], prefix=["mnth"], drop_first=True)
dummy3 = pd.get_dummies(test_d, columns=["weathersit"], prefix=["weathersit"], drop_first=True)
dummy1 = dummy1.drop(columns=dummy1.columns[0:2])
dummy2 = dummy2.drop(columns=dummy2.columns[0:2])
dummy3 = dummy3.drop(columns=dummy3.columns[0:2])

df=pd.concat([df,dummy1,dummy2,dummy3], axis=1)
```

### model

```python
# 不要な説明変数を削除
x, y = df.drop("cnt", axis=1), df["cnt"]
# 訓練⽤データセットとテスト⽤データセットへの分割
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=0)

from sklearn.linear_model import LinearRegression
# モデルの定義
clf_model = LinearRegression()

# モデルの訓練
clf_model.fit(x_train, y_train) # 訓練データセットの説明変数と⽬的変数を引数に渡す
# LinearRegression(copy_X=True, fit_intercept=True, n_jobs=None, normalize=False)

stem_df = pd.DataFrame({"features":x_train.columns,"coef":clf_model.coef_}).sort_values(by="coef")
display(stem_df)
print("切⽚(バイアス)は{}".format(clf_model.intercept_))
```

```python
# テストデータセットで推論
test_pre_lr = clf_model.predict(x_test)
# 予測した結果と実際の値を結合する
test_result_lr = pd.DataFrame({"actual": y_test, "predict":test_pre_lr}).reset_index(drop=True)
# 訓練用データセットで推論
train_pre_lr = clf_model.predict(x_train)
# 予測した結果と実際の値を結合する
train_result_lr = pd.DataFrame({"actual": y_train, "predict":train_pre_lr}).reset_index(drop=True)

from sklearn.metrics import mean_absolute_error #MAEを算出するためのライブラリ
train_mae_lr = mean_absolute_error(y_train, train_pre_lr)
print("訓練データセットのMAE: {0}".format(train_mae_lr))

test_mae_lr = mean_absolute_error(y_test, test_pre_lr)
print("テストデータセットのMAE: {0}".format(test_mae_lr))
```

## LightGBM

```python
from lightgbm import LGBMRegressor

# 使用するチューニング対象外のパラメータ
params = {
    'objective': 'regression',  # 最小化させるべき損失関数
    'metric': 'rmse',  # 学習時に使用する評価指標(early_stoppingの評価指標にも同じ値が使用される)
    'random_state': 0,  # 乱数シード
    "learning_rate": 0.1,
    'boosting_type': 'gbdt',
    'n_estimators': 10000,  # 最大学習サイクル数。early_stopping使用時は大きな値を入力
    'verbose': -1,  # これを指定しないと`No further splits with positive gain, best gain: -inf`というWarningが表示される
    'early_stopping_round': 10  # ここでearly_stoppingを指定
    }

# モデル作成
model = LGBMRegressor(**params)

# モデルの訓練
eval_set = [(x_test, y_test)]  # 修正：検証データセットを指定
model.fit(x_train, y_train, eval_set=eval_set, eval_metric='rmse')  # 修正：eval_setとeval_metricを指定

# テストデータセットで推論
test_pre_lr = model.predict(x_test)
# 予測した結果と実際の値を結合する
test_result_lr = pd.DataFrame({"actual": y_test, "predict":test_pre_lr}).reset_index(drop=True)
# 訓練用データセットで推論
train_pre_lr = model.predict(x_train)
# 予測した結果と実際の値を結合する
train_result_lr = pd.DataFrame({"actual": y_train, "predict":train_pre_lr}).reset_index(drop=True)

from sklearn.metrics import mean_absolute_error #MAEを算出するためのライブラリ
train_mae_lr = mean_absolute_error(y_train, train_pre_lr)
print("訓練データセットのMAE: {0}".format(train_mae_lr))

test_mae_lr = mean_absolute_error(y_test, test_pre_lr)
print("テストデータセットのMAE: {0}".format(test_mae_lr))
```

```python
plt.subplots(figsize=(15,10))
x = test_result_lr["actual"]
y = test_result_lr["predict"]
plt.scatter(x, y)
plt.plot([0, 8800], [0, 8800], '-')
plt.xlabel("actual")
plt.ylabel("predict")
plt.xlim(0,8800)
plt.ylim(0,8800)
plt.grid()
plt.show()
```
